```python
from pathlib import Path
import json
```

```python
def json_to_csv(in_file_path, out_file_path):
    metric_names = [
        #'mean_solving_steps', 
        'initial_state_accuracy_micro', 'initial_state_accuracy_macro',
        'end_state_accuracy_micro', 'end_state_accuracy_macro'
    ]

    with open(in_file_path, 'r') as input, open(out_file_path, 'w') as output:
        output.write(';'.join(['name'] + metric_names)+'\n')
        data: dict = json.load(input)
        for name, metrics in data.items():
            output.write(';'.join(
                [name] 
                + [str(metrics[metric_name]) for metric_name in metric_names])+'\n')

```

```python
for path_item in Path('').iterdir():
    if path_item.is_file():
        file_name = path_item.name
        name, file_type = file_name.split('.')
        if file_type.lower() == 'json':
            json_to_csv(path_item, path_item.parent/f'{name}.csv')
            print(file_name, name, file_type)
```
